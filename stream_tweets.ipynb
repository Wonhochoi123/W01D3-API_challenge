```python
limit=1000
#who uses ':)' or ':(' anymore these days? everyone uses emojis!!!
happy=[':)', 'XD','üòÄ','üòÉ','üòÑ','üòÅ','üòÜ','ü§£','üôÇ','üôÉ','üòâ','üòä','üòá','ü•∞','üòç','ü§©','üòò','üòó']
unhappy=[':(','üòï','üòü','üôÅ','‚òπÔ∏è','üò∞','üò•','üò¢','üò≠','üò±','üòñ','üò£','üòì','üòû','ü§¨','üò°','üò†','üò´','üò©']
def main(path, FILTER, languages):
    count=0
    with open(path, 'a') as f:
        
        for line in api.GetStreamFilter(track=FILTER, languages=languages):
            tweet=line['text']
            if count>limit:
                return
            else:
                if bool([ele for ele in happy if(ele in tweet)]):
                    f.write('happy ')
                    f.write(json.dumps(line))
                    f.write('\n')
                    f.write('\n')
                    count+=1
                    print(line)
                if bool([ele for ele in unhappy if(ele in tweet)]):
                    f.write('unhappy ')
                    f.write(json.dumps(line))
                    f.write('\n')
                    f.write('\n')
                    count+=1
                    print(line)
```

```python
main("twitter_feeling.txt",languages=['en'], FILTER= ['#datascience'])
```
